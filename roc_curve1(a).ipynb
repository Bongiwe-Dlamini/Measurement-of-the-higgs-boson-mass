```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated data (replace with your data)
signal_data = np.random.normal(128, 1, 986)
background_data = np.random.normal(120, 67, 36169 )

cut_values = np.arange(120, 130, 1)
signal_efficiency = []
background_rejection = []

for cut in cut_values:
    signal_pass = signal_data[signal_data > cut]
    background_pass = background_data[background_data > cut]
    signal_eff = len(signal_pass) / len(signal_data)
    background_rej = 1 - (len(background_pass) / len(background_data))
    signal_efficiency.append(signal_eff)
    background_rejection.append(background_rej)

plt.figure()
plt.plot(signal_efficiency, background_rejection, marker='o', linestyle='-', color='b')
plt.xlabel('Signal Efficiency')
plt.ylabel('Background Rejection')
plt.title('Signal Efficiency vs. Background Rejection')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.savefig("signal_eff_vs_bkg_rej.png")
plt.show()
```
